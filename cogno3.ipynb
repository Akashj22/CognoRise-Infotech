```python
dfe=pd.get_dummies(df,columns=['Sex','Embarked'],drop_first=True,dtype=int)
dfe
```

```python
dfe.drop(['PassengerId','Name','Ticket','Fare','Cabin'],axis=1,inplace=True)
dfe
```

```python
dfe['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
dfe.isna().sum()
```

```python
x=dfe.drop(['Survived'],axis=1).values
x
```

```python
y=dfe['Survived'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

knn=KNeighborsClassifier()
nav=GaussianNB()
sv=SVC()

lst=[knn,nav,sv]

for i in lst:
  print('model started :',i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('confusion metrics is')
  print(confusion_matrix(y_test,y_pred))
  print('accuracy score is',accuracy_score(y_test,y_pred))
  print('classification report',classification_report(y_test,y_pred))
  print('model ended')
  print('-'*100)
```
